```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

```python
URL = 'https://www.cricbuzz.com/cricket-team/india/2/players'
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
```

```python
container = soup.find('div', attrs={'id' : 'series-news-list'})
```

```python
container = container.findChild()
```

```python
playerLinks = container.find_all('a')
```

```python
linksList = []
for link in playerLinks:
    playerName = link['title']
    playerLink = 'http://cricbuzz.com' + link['href']
    linksList.append([playerName, playerLink])
```

```python
df = pd.DataFrame(linksList, columns=['Name', 'Link'])
```

```python
df.to_csv('Player Links.csv')
```

```python
soup.find('div', attrs={'id' : 'Players'})
```

```python
searchQuery = ['Suryakumar yadav',
'Ishan kishan',
'Krunal Pandya',
'Washington Sundar',
"Deepak Chahar ",
'T. Natarajan',
'Varun chakravarthy',
'Rahul chahar']
```

```python
playerLinks = []
for name in searchQuery:
    driver = webdriver.PhantomJS()
    URL = 'https://www.cricbuzz.com/search?q=' + name
    driver.get(URL)
    p_element = driver.find_element_by_xpath('//*[@id="Players"]/div[2]/div/div/div[2]/a')
    playerLinks.append(p_element.get_attribute('href'))
```
